```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

# Part-1: Data Exploration and Pre-processing

```python
# 1) load the given dataset 

data = pd.read_excel("1768299714153-ML Project - KNN Loan Applicant Credit Risk Analysis U 16955473070.xlsx")
data.head()
```

```python
# 2) Check for the null values 

data.isnull().sum()
```

```python
# 3) Get basic information from data

print(data.columns)
print(data.dtypes)
print(data.info())
```

```python
# 4) Describe the dataset 

data.describe()
```

```python
# 5) Display scatterplot between age & Total work Experience

import matplotlib.pyplot as plt
plt.scatter(data["Age"], data["Total Work Experience"])
plt.title("Scatterplot Between Age & Total Work Experience")
plt.ylabel("Total Work Experience")
plt.xlabel("Age")
plt.show()
```

```python
# 6) Display box plot for age 

plt.boxplot(data["Age"])
plt.title("Box Plot For Age")
plt.ylabel("Age")
plt.show()
```

```python
# 7) Display box plot for Cibil score 

plt.boxplot(data["Cibil score"])
plt.title("Box Plot For Cibil score")
plt.ylabel("Cibil score")
plt.show()
```

```python
# 8) Create target and features data where target is Total bounces past 12 months

x = data.drop("Total bounces past12months", axis = 1)
y= data["Total bounces past12months"]
```

# Part-2: Working with Model

```python
# 1) Split data into training and testing sets 

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
# 2) Create a KNN classifier between Features and target data
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
# 3) Display the test score 
test_score = knn.score(x_test, y_test)
print("Test Score:", test_score)
```

```python
# 4) Display the training score
# Training score (training data par performance)
train_score = knn.score(x_train, y_train)
print("Training Score:", train_score)
```

```python
# 5) Print the accuracy score 

from sklearn.metrics import accuracy_score

y_pred = knn.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```

```python
# 6) Try 1 to 14 k values for classifier

from sklearn.neighbors import KNeighborsClassifier

for k in range(1, 15):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    
    print(f"K = {k} model trained")
```

```python
# 7) Display training and testing score for all the 1 to 14 k values

from sklearn.neighbors import KNeighborsClassifier

for k in range(1, 15):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    
    train_score = knn.score(x_train, y_train)
    test_score = knn.score(x_test, y_test)
    
    print(f"K = {k} | Train Score = {train_score:.3f} | Test Score = {test_score:.3f}")
```
